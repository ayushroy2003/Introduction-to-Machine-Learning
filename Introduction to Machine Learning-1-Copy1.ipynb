Q5: What are the main differences between supervised, unsupervised, and semi-supervised learning?

Ans-  

1. Supervised learning

Uses : Fully labeled data

Example : Email spam detection

2. Unsupervised Learning
Uses : No labeled data


Goal : Find patterns or groupings in data


Example : Customer Segementation

3. Semi-Supervised Learning 
Uses : A mix of small labeled set and a large unlabled set


Goal : Impove learning by leveraging some labled data


Example : Image classification with limited labled images


Q6: What is train, test and validation split? Explain the importance of each term.

Ans-  

### Training Set : 
Definition : THe Portion of the dataset used to train the model and help it learn patterns.

Imortance : It is where the model learns the relationships between the input features and the outputs 

### Validation Set :

Definition : A subset of data used to tune the model  hyperparameter and evalute its perfomance during training 

Importance : Helps prevent overfitting Overfitting by allowing you to test the models performance on unseen data while adjusting parameter for better results 

### Test set 

Defination : A separate portion of the dataset that the model has never seen during training or validation 

Importance : Provide and unbiased assesment of the model's final perfoemance and generalization to new data 



Q7: How can unsupervised learning be used in anomaly detection?

Ans- Unsupervised learning can be highly  effective for anomaly detection because it doesn't require labeled data and can learn pattern in data to identify What is "normal' .

### How it work 

- Pattern Learning : The model is trained on a dataset where most of the data points are normal , and it learns the common patterns or structure

- Anomaly Indentification : Once the model understand what "normal' data looks like , it can flag any data points that devigate Significantly from these patterns as anomalies
